```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hero = pd.read_csv('sample_101_Mlbb_Heroes.csv')

hero.head()
```

```python
hero.info()
hero.describe()
```

```python
hero = hero.drop(
    columns=["Secondary_Role", "Title", "Voice_Line", "Release_Date"],
    errors="ignore"
)
```

```python
hero.head()
```

```python
hero.isnull().sum()
```

```python
hero.dtypes
```

```python
hero['Total_damage'] = hero['Phy_Damage'] + hero['Mag_Damage']
```

```python
hero['Total_defence'] = hero['Phy_Defence'] + hero['Mag_Defence']
```

```python
hero['Win_rate'] = hero["Esport_Wins"] / (
    hero["Esport_Wins"] + hero["Esport_Loss"]
)
```

```python
hero['Win_rate'].describe()
```

```python
hero.duplicated().sum()
```

```python
hero.head()
```

```python
hero['Overall_score'] = (hero['Total_damage'] + hero['Total_defence'] + hero['Mov_Speed'])
```

```python
hero.head()
```

```python
plt.figure()
plt.hist(hero["Win_rate"], bins=10)
plt.xlabel("Win Rate")
plt.ylabel("Number of Heroes")
plt.title("Distribution of Hero Win Rate")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x="Primary_Role",y="Win_rate",data=hero)
plt.xlabel("Primary Role")
plt.ylabel("Win Rate")
plt.title("Win Rate Distribution by Hero Role")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.scatter(hero["Overall_score"], hero["Win_rate"])
plt.xlabel("Overall Score")
plt.ylabel("Win Rate")
plt.title("Overall Score vs Win Rate")
plt.show()
```

```python
corr = hero[["Win_rate", "Total_damage", "Total_defence", "Mov_Speed"]].corr()
```

```python
plt.figure()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
X = hero[['Overall_score']]
y = hero['Win_rate']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=12)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred) 

print("MAE:", mae)
print("RMSE:", rmse)
print("R2:", r2)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         linestyle="--")
plt.xlabel("Actual Win Rate")
plt.ylabel("Predicted Win Rate")
plt.title("Predicted vs Actual Win Rate")
plt.tight_layout()
plt.show()
```

```python
import joblib
joblib.dump(model, "draft.joblib")
print("Model saved as draft.joblib")
```
