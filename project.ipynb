```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hero = pd.read_csv('sample_101_Mlbb_Heroes.csv')

hero.head()
```

```python
hero.info()
hero.describe()
```

```python
hero = hero.drop(
    columns=["Secondary_Role", "Title", "Voice_Line", "Release_Date"],
    errors="ignore"
)
```

```python
hero.head()
```

```python
hero.isnull().sum()
```

```python
hero.dtypes
```

```python
hero['Total_damage'] = hero['Phy_Damage'] + hero['Mag_Damage']
```

```python
hero['Total_defence'] = hero['Phy_Defence'] + hero['Mag_Defence']
```

```python
hero['Win_rate'] = hero["Esport_Wins"] / (
    hero["Esport_Wins"] + hero["Esport_Loss"]
)
```

```python
hero['Win_rate'].describe()
```

```python
hero.duplicated().sum()
```

```python
hero.head()
```

```python
hero_eda = hero[["Total_damage", "Total_defence", "Mov_Speed", "Win_rate"]]
```

```python
hero_eda.describe()
```

```python
corr = hero_eda.corr()
print(corr)
```

```python
plt.figure()
plt.hist(hero['Win_rate'],bins=10)
plt.xlabel('Win Rate')
plt.ylabel('Number of Heroes')
plt.title('Distribution of Hero Win Rate')
plt.show()
```

```python
plt.figure()
plt.scatter(hero['Total_damage'], hero['Win_rate'])
plt.xlabel('Total Damage')
plt.ylabel('Win Rate')
plt.title('Total Damage vs Win Rate')
plt.show()
```

```python
plt.figure()
plt.scatter(hero['Total_defence'], hero['Win_rate'])
plt.xlabel('Total Defence')
plt.ylabel('Win Rate')
plt.title('Total Defence vs Win Rate')
plt.show()
```

```python
plt.figure()
plt.scatter(hero['Mov_Speed'], hero['Win_rate'])
plt.xlabel('Movement Speed')
plt.ylabel('Win Rate')
plt.title('Movement Speed vs Win Rate')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)
plt.title('Correlation Heatmap of Hero Attributes')
plt.show()
```

```python
role = hero.groupby('Primary_Role')['Win_rate'].mean().sort_values(ascending=False)

plt.figure()
role.plot(kind='bar')
plt.xlabel('Primary Role')
plt.ylabel('Average Win Rate')
plt.title('Average Win Rate by Role')
plt.show()
```

```python
X = hero[['Total_damage','Total_defence','Mov_Speed']]
y = hero['Win_rate']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=12)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
import joblib
joblib.dump(model, "draft.joblib")
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print('Linear Regression R:', r2)
print('Linear Regression MSE:', mse)
```
